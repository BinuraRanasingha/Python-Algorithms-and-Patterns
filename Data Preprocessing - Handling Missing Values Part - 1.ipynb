# Dropping missing values

```python
data.dropna(axis=1)
```

```python
#Checking for more null values
(data.isna().sum() / len(data)) * 100
```

**thresh parameter - here we keep only columns where there are at least the 80% of not null values**

```python
data.dropna(thresh=0.8 * len(data),axis=1,inplace=True)
```

```python
(data.isna().sum()/len(data)) * 100
```

# Replacing the missing values

```python
data.dtypes
```

## Numerical values

```python
numeric = data.select_dtypes(include=np.number)
numeric_columns = numeric.columns
```

Filling the NaN values with the average value, given by the mean() function

```python
data[numeric_columns] = data[numeric_columns].fillna(data.mean())
```

```python
(data.isna().sum()/len(data)) *100
```

## Categorical values

The dtypes the categorial columns are described as objects. Can select the object columns.Would like to consider only boolean columns. However the object type includes also the column class, which is a string so select all the object columns, and then  remove them from the column class so can convert the type of the result to bool.

```python
boolean_columns = data.select_dtypes(include=np.object).columns.tolist()
boolean_columns.remove('class')
data[boolean_columns] = data[boolean_columns].astype('bool')
```

```python
#Replacing the missing values with the mode
data[boolean_columns].fillna(data.mode())
```

```python
(data.isna().sum()/len(data)) *100
```

## Interpolation

Linear interpolation is the technique of determining the values of the functions of any intermediate points when the values of the 2 adjacent points are known. Linear interpolation is basically the estimation of an unknown values that falls within 2 known values.

```python
df = pd.read_csv('D:\\SLIIT\\3rd year 2nd sem\\Fundamentals of Data Mining\\Coding\\Hepatitis.csv')
df
```

```python
(df.isna().sum()/len(df)) * 100
```

```python
numeric = df.select_dtypes(include=np.number)
numeric_columns = numeric.columns
```

```python
df.head(10)
```

```python
df[numeric_columns] = df[numeric_columns].interpolate(method='linear',limit_direction='forward')
```
