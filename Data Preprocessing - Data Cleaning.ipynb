```python
#Importing the necessary libraries 
import pandas as pd
from pandas_profiling import ProfileReport
```

```python
#Loading the dataset
data = pd.read_excel("D:\\SLIIT\\3rd year 2nd sem\\Fundamentals of Data Mining\\Coding\\rainfall.xlsx",skiprows=2)
data 
```

```python
#Removing additional columns
data = pd.read_excel("D:\\SLIIT\\3rd year 2nd sem\\Fundamentals of Data Mining\\Coding\\rainfall.xlsx",skiprows=2,usecols='B:D')
data
```

```python
#Dropping null columns
data.dropna(inplace=True,axis=0)
data
```

```python
#Splitting the Month,period column
splitted_columns=data['Month, period'].str.split(',',expand=True)
splitted_columns
```

```python
data['Month'] = splitted_columns[0]
data['Period'] = splitted_columns[1]
data
```

```python
data.drop('Month, period',axis=1,inplace=True)
data.head()
```

```python
def remove_mm(x):
    if type(x) is str:
        return x.replace('mm','')
    else:
        return x
```

```python
data['Lake Victoria'] = data['Lake Victoria'].apply(lambda x: remove_mm(x))
data['Simiyu'] = data['Simiyu'].apply(lambda x:remove_mm(x))
data.head()
```

```python
data['Lake Victoria'] = pd.to_numeric(data["Lake Victoria"])
data['Simiyu'] = pd.to_numeric(data["Simiyu"])
```
